```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import csv

%matplotlib notebook
```

```python
#read csv file with encoding set to 'latin-1' because of polish language
df_14 = pd.read_csv('FRKF-inwestycje-2014.csv', sep='|' , encoding='latin-1')
df_15 = pd.read_csv('FRKF-inwestycje-20154kw.csv', sep='|' , encoding='latin-1')
df_16 = pd.read_csv('umowyFRKF2016.csv', sep='|' , encoding='latin-1')
```

```python
#check how the data is formatted for cleaning
#actually the data is comming into one column from the reading csv function
#df_14.iloc[4, 0]
#df_15.iloc[5 0]
#df_16.iloc[471, 0]
```

```python
#clean data and retrieve data for 2014
#province names
provinces_2014 = []
#amount spent by row
amount_2014 = []

for i in range (len(df_14)):
    line = str(df_14.iloc[i, 0])
    provinces_2014.append(line.split(';')[1])
    amount_2014.append(line.split(';')[-1])

#amount coming in the form with name at the end, so it has to be removed
amount_14_1 = []
for i in range(len(amount_2014)):
    amount_14_1.append(amount_2014[i].split(',')[0])

#remove the space which is as a separator in numbers
cleaned_amount_14 = []
for i in range(len(amount_14_1)):
    item = amount_14_1[i].split(' ')
    cleaned_amount_14.append("".join(item))

#finally update the dataframe for 2014
df_14 = pd.DataFrame({'Provinces': provinces_2014,
                     'Amount_Invested': cleaned_amount_14},
                     index = [x for x in range(1, len(provinces_2014) + 1)])

#replace provinces name bto english ones
df_14 = df_14.replace({'Provinces' : 
                       {'dolnol¹skie': 'Lower Silesia', 'kujawsko-pomorskie': 'Kujawsko-Pomorskie',
                        'lubelskie': 'Lublin', 'lubuskie': 'Lubuskie', '³ódzkie': 'Łódź',
                       'ma³opolskie': 'Małopolska', 'mazowieckie': 'Masovia', 'opolskie': 'Opole',
                       'podkarpackie': 'Podkarpackie', 'podlaskie': 'Podlasie', 'pomorskie': 'Pomerania',
                       'l¹skie' : 'Silesia', 'wiêtokrzyskie': 'Holy Cross' , 
                        'warmiñsko-mazurskie' : 'Warmia-Masuria', 'wielkopolskie': 'Wielkopolska',
                       'zachodniopomorskie': 'West Pomerania'}})

#change the format of the amount to numeric
df_14['Amount_Invested'] = pd.to_numeric(df_14['Amount_Invested'])
```

```python
#add a column sum to every dataframe. the sum is returned by groupby function by provinces
df_14['SUM_2014'] = df_14['Amount_Invested'].groupby(df_14['Provinces']).transform('sum')
df_15['SUM_2015'] = df_15['Amount_Invested'].groupby(df_15['Provinces']).transform('sum')
df_16['SUM_2016'] = df_16['Amount_Invested'].groupby(df_16['Provinces']).transform('sum')
```

```python
#just take the columns provinces and sum (created above)
df_14 = df_14[['Provinces', 'SUM_2014']]
df_15 = df_15[['Provinces', 'SUM_2015']]
df_16 = df_16[['Provinces', 'SUM_2016']]
```

```python
#return only one column for every province returned by the groupby above
df_14 = df_14.groupby('Provinces').apply(lambda x: x.iloc[[0]])
df_15 = df_15.groupby('Provinces').apply(lambda x: x.iloc[[0]])
df_16 = df_16.groupby('Provinces').apply(lambda x: x.iloc[[0]])
```

```python
#drop the unnecessary columns from the index
df_14.index = df_14.index.droplevel(0)
df_15.index = df_15.index.droplevel(0)
df_16.index = df_16.index.droplevel(0)
```

```python
#sort values by provinces in order to have the same order in every dataframe
df_14 = df_14.sort_values(by = ['Provinces'])
df_15 = df_15.sort_values(by = ['Provinces'])
df_16 = df_16.sort_values(by = ['Provinces'])
```

```python
#convert the data to plot to list for better plotting
list_provinces = df_14['Provinces'].to_list()
sum_2014 = df_14['SUM_2014'].to_list()
sum_2015 = df_15['SUM_2015'].to_list()
sum_2016 = df_16['SUM_2016'].to_list()
```

```python
#locale will help to display the numbers with a separator
import locale
locale.setlocale(locale.LC_ALL, '')

#width of a bar
width = 0.28
ind = np.arange(len(list_provinces))

#start plotting
fig, ax = plt.subplots(figsize=(12, 15))
fig.subplots_adjust(left = 0.3, right = 0.80, top = 0.9, bottom = 0.1)

#first plot with values from 2014
plot1 = ax.barh(ind, sum_2014, width, label = '2014')
#add label to the bars created
for i, v in enumerate(sum_2014):
    ax.text( v + 4, i - 0.1, f'{v:n}' , color = 'blue', fontsize = 11, fontweight = 'bold')

#second plot with values from 2015
plot2 = ax.barh(ind + width, sum_2015, width, label = '2015 (4th quarter)', color='orange')
#add label to the bars created
for i, v in enumerate(sum_2015):
    ax.text( v + 4, i + 0.2, f'{v:n}', color = 'orange', fontsize = 11, fontweight = 'bold')

#third ploy with values from 2016
plot3 = ax.barh(ind + (2*width), sum_2016, width, label = '2016', color='green')
#add label to the bars created
for i, v in enumerate(sum_2016):
    ax.text( v + 4, i + 0.5, f'{v:n}', color = 'green', fontsize = 11, fontweight = 'bold')

#remove the unnecessary spines (set them to invisible)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

#name the axis and just align them
ax.set_xlabel('Amount (in 10,000,000 zł)', fontweight = 'bold', fontsize= 15)
ax.set_ylabel('Names provinces', rotation = 90, fontweight = 'bold', fontsize= 15)
plt.suptitle('Investment in Sport by the Physical Culture Development funds \n In each province \n Amount in zlotych (1$ = 3.78 zł) \n', fontweight= 'bold', fontsize = 16)
ax.set_title('Poland (2014 - 2015(4th quarter) - 2016) ', fontsize = 16, pad= 3)
ax.set_yticks(ind)
ax.set_yticklabels(list_provinces, fontsize = 14)
ax.tick_params(axis=u'both', which = u'both', length=0)
#show the legend
plt.legend(bbox_to_anchor=(1, 1), loc=1, borderaxespad=0)
#finally save the figure created
plt.savefig('compare.jpeg')
```
